```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#reading the dataset
books=pd.read_csv('Books.zip',encoding ='unicode_escape',error_bad_lines=False)
books
```

```python
#combining the columns of main features
features=[]
for i in range(0,books.shape[0]):
  features.append(str(books['bookTitle'][i])+' , '+str(books['bookAuthors'][i]))
books['combined_features']=features
books
```

```python
#using countvectorizer from sklearn lib to transform a given text into a vector on the basis of the frequency (count) of each word that occurs in the entire text.
cm=CountVectorizer().fit_transform(books["combined_features"])
cs=cosine_similarity(cm) #Cosine similarity measures the similarity between two vectors of an inner product space.
print(cs)
```

```python
#giving an input(book name) and finding the index of the input in dataset.

Title='Pride and Prejudice'
c=0
def index_of_Title(Title):
    for i in range(0,books.shape[0]):
        if(Title==books['bookTitle'][i]):
            c=i
            break
    return c

index_of_Title(Title)
```

```python
#Taking cosine similarity matrix of the index of the input entered.

def similar_matrix(cs):
    c=index_of_Title(Title)
    result=list(cs[c])
    sort_index = np.argsort(result)
    p=list(sort_index)
    p.reverse()
    return p
```

```python
#recommending the books based on the cosine similarity matrix of the entered input i.e recommending books with highest cosine similarity value of that index.

def recommend(Title):
    b = similar_matrix(cs)
    f=0
    for i in b:
        if i!=b[0]:
            if(f<5):
                print(books['bookTitle'][i])
                f=f+1
                
recommend(Title)
```

```python
#sorting dataset based on the rating column
sorted_ratings=books.sort_values(['bookRating'], ascending=[False])

#sorting the indexes dataset based on the rating column
sorted_ratings= sorted_ratings.reset_index(drop=True)

sorted_ratings
```

```python
#recommending high rated books
def sorted_rating(sorted_ratings):
    k=[]
    t=0
    for i in sorted_ratings['bookTitle']:
        if(t<5):
            k.append(i)
            t=t+1
            
    return k

sorted_rating(sorted_ratings)
```
